```python
import pandas as pd
```

```python
props_2017 = pd.read_csv('/../../../Data/properati-AR-2017-08-01-properties-sell-six_months.csv')
props_2017[['created_on']] = pd.to_datetime(props_2017['created_on'],format='%Y-%m-%d')
props_2017.head()
```

```python
props_2017.drop(['id','operation','place_with_parent_names','created_on','lat-lon','country_name','geonames_id','title','operation','image_thumbnail','properati_url','description'], axis = 1, inplace = True)
```

```python
#Rows que no tengan ni price, ni price_aprox_usd, price_aprox_local_currency , ni surface_total_in_m2, ni surface_covered_in_m2
props_2017_filtered = props_2017.loc[(props_2017['price'].notnull()) & (props_2017['price_aprox_usd'].notnull()) & \
               (props_2017['price_aprox_local_currency'].notnull()) & (props_2017['surface_total_in_m2'].notnull()) & (props_2017['surface_covered_in_m2'].notnull())]
props_2017_filtered.head()
```

```python
#Rows que tengan valores de precios erroneos
props_2017_filtered.describe()
```

```python
#Genero el csv filtrado
props_2017_filtered.to_csv('props_2017_test.csv',sep=';',index=False)
```
